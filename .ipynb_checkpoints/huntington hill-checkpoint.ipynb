```python
import plotly
import pandas as pd
import geojson
import geopandas as gpd
import plotly.express as px
pd.options.mode.chained_assignment = None 
from geojson_rewind import rewind
import apportionpy.apportionment as ap

df_2019 = pd.read_csv('Final_Data/2019/2019.csv')
df_2014 = pd.read_csv('Final_Data/2014/2014.csv')
df_2009 = pd.read_csv('Final_Data/2009/2009.csv')
df_2004 = pd.read_csv('Final_Data/2004/2004.csv')
df_1999 = pd.read_csv('Final_Data/1999/1999.csv')
df_1996 = pd.read_csv('Final_Data/1996/1996.csv')
df_1991 = pd.read_csv('Final_Data/1991/1991.csv')
df_1989 = pd.read_csv('Final_Data/1989/1989.csv')
df_1984 = pd.read_csv('Final_Data/1984/1984.csv')
df_1980 = pd.read_csv('Final_Data/1980/1980.csv')
df_1977 = pd.read_csv('Final_Data/1977/1977.csv')
df_1971 = pd.read_csv('Final_Data/1971/1971.csv')
df_1967 = pd.read_csv('Final_Data/1967/1967.csv')
df_list = []
df_list.append(df_2019)
df_list.append(df_2014)
df_list.append(df_2009)
df_list.append(df_2004)
df_list.append(df_1999)
df_list.append(df_1996)
df_list.append(df_1991)
df_list.append(df_1989)
df_list.append(df_1984)
df_list.append(df_1980)
df_list.append(df_1977)
df_list.append(df_1971)
df_list.append(df_1967)




i = 0
for df_ in df_list:
    df_ = df_.fillna(0)
    seats = df_.loc[len(df_.index)-1].at['Seats']
    populations = df_['Electors'][:-1]
    method_1 = 'huntington hill'
    result_1 = ap.Apportion(seats=seats, populations=populations, method=method_1)
    x = result_1.fair_shares
    sum_ = sum(x)
    x.append(sum_)
    df_['huntington hill'] = x
    #df_['huntington hill'] = result_2.fair_shares
    df_['Difference'] = df_['huntington hill'] - df_['Quota']
    df_list[i] = df_
    i += 1
    
    
    

def hover_text(df):
    df['text'] = ""
    for ind in df.index:
        df['text'][ind] =  'huntington hill Allocation : ' + str(df['huntington hill'][ind]) + "         Quota:" + str(df['Quota'][ind]) + "    Electors:" + str(df['Electors'][ind]) 
        
for x in df_list:
    hover_text(x)
    

geojson_list = []
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/2019.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/2014.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/2009.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/2004.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1999.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1996.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1991.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1989.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1984.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1980.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1977.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1971.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)
with open('/Users/sunidhisingh/Desktop/Coursework/COD300/Code/Geojson/1967.json') as f:
    gj = geojson.load(f)
    gj = rewind(gj, rfc7946=False)
geojson_list.append(gj)



data = [dict(type='choropleth', 
                    locations = df_list[0]['State/UT'].astype(str),
                    z = df_list[0]['Difference'].astype(str),
                     geojson = geojson_list[0],
                     featureidkey='properties.ST_NM',
                     text = df_list[0]['text']
                     
            )]

for i in range(1,len(df_list)):
    data.append(data[0].copy())
    data[-1]['z'] = df_list[i]['Difference'].astype(str)
    data[-1]['locations'] = df_list[i]['State/UT'].astype(str)
    data[-1]['text'] = df_list[i]['text']
    data[-1]['geojson'] = geojson_list[i]
    
    
years = [2019, 2014, 2009, 2004 , 1999 , 1996 , 1991 , 1989 , 1984 , 1980 , 1977 , 1971 , 1967 ]

steps = []

for i in range(len(data)):
	step = dict(method='restyle',
				args=['visible', [False] * len(data)],
				label='Year {}'.format(years[i]))
	step['args'][1][i] = True
	steps.append(step)

sliders = [dict(active=0, pad={"t": 1}, steps=steps)]

layout = dict(sliders=sliders,
             geo = dict(fitbounds="locations", visible=False))

fig = dict(data=data, 
           layout=layout)


plotly.offline.iplot(fig)
plotly.offline.plot(fig, filename='huntington hill.html')
```
